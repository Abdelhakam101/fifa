**What is the average Age ?**

```python
df['age'].mean()
```

**What is the greatest Height? and name of the player with that height?**

```python
# greatest height
# df['height'].max()
df.sort_values(by='height',ascending=False).head(1)
```

```python
# name of the player
df.sort_values(by='height',ascending=False).head(1)[['name']]
```

**What is the smallest weight? and name of the player with that weight? and for which club plays?**

```python
# df['weight'].min()
df.sort_values(by='weight',ascending=True).head(1)
```

```python
# name of the player
df.sort_values(by='weight',ascending=True)[['name','club']].head(1)
```

**What is the position of Cristiano Ronaldo?**

```python
df[df['name'] == 'Cristiano Ronaldo'][['name','position']]
```

**What was the average (mean) Value of all players per year(Joined column) ?**

```python
df['value'].isna().sum()
```

```python
df.groupby('joined')[['value']].sum().reset_index()
```

```python
df.groupby('joined')[['value']].sum().sort_values(by='value',ascending = False).reset_index()
```

**get Number of players for each club**

```python
df.groupby('club')[['name']].count().reset_index()
```

**How many unique Clubs are there?**

```python
df['club'].unique()
```

```python
df['club'].nunique()
```

**What are the top 5 most common positions?**

```python
df.groupby('position')[['name']].count().sort_values(by='name',ascending = False).head(5).reset_index()
```

**What are the top 5 most common nationality?**

```python
df.groupby('nationality')[['name']].count().sort_values(by='name',ascending = False).head(5).reset_index()
```

**What are the names of players has Egyptian Nationality?**

```python
df[['name']][df['nationality'] == 'Egypt'].reset_index()
```

**How many players joined in 2018 with skill moves=2?** 

```python
len(df[(df['joined' ] == 2018 ) & (df['skill_moves'] == 2 )])
```

**How many players whom name starts with letter  'M'? (This is pretty tricky)**

```python
len(df[df['name'].str.startswith('M')])
```

```python
len(df[df['name'].str[0] == 'M'])
```
